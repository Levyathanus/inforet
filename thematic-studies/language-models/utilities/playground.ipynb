```python
import numpy as np
import pandas as pd
from tqdm import tqdm_notebook
```

```python
f = {
    'indentity function': lambda x: x,
    'sign function': lambda x: [np.sign(y) for y in x],
    'sigmoid function': lambda x: [1 / (1 + np.exp(-y)) for y in x],
    'tanh function': lambda x: [(np.exp(2*y) - 1) / (np.exp(2*y) + 1) for y in x],
    'ReLU': lambda x: [np.max([y, 0]) for y in x],
    'hard tanh': lambda x: [np.max([np.min([y, 1]), -1]) for y in x]
}
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
with plt.style.context('seaborn-darkgrid'):
    x = np.arange(-5, 6, 0.1)
    fig, ax = plt.subplots(figsize=(6.5, 8), nrows=len(f)//2, ncols=2)
    r, c = 0, 0
    for i, (n, fun) in enumerate(f.items()):
        ax[r, c].set_title(n)
        ax[r, c].plot(x, fun(x))
        ax[r, c].xaxis.set_major_locator(plticker.MultipleLocator(1))
        if i >= 2:
            r = i + 1 - 3
            c = 1
        else:
            r = i + 1
    plt.tight_layout()
    plt.savefig(
        '/Users/alfio/Teaching/2019-20/inforet/lectures/thematic-studies/S1-language-models/imgs/activation.png')
    plt.show()
```
