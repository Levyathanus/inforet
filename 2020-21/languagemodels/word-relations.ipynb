```python
from nltk.tokenize import sent_tokenize
from nltk import word_tokenize
```

```python
def skip(sequence, s=2):
    n = 2
    k_grams = []
    for i in range(len(sequence)):
        for j in range(i+1, min(i+s, len(sequence))):
            k_grams.append((sequence[i], sequence[j]))
    return k_grams
```

```python
def tokenize(text):
    sequences = []
    sentences = sent_tokenize(text.lower())
    for sentence in sentences:
        tokens = word_tokenize(sentence)
        sequence = skip(tokens, s=4)
        yield sequence
```

```python
from collections import defaultdict
```

```python
M = defaultdict(lambda: defaultdict(lambda: 0))
for document in data_train.data[:100]:
    for sequence in tokenize(document):
        for a, b in sequence:
            M[a][b] += 1
```

```python
dictionary = list(M.keys())
X = np.zeros((len(dictionary), len(dictionary)))
for k, v in M.items():
    for z, c in v.items():
        try:
            X[dictionary.index(k)][dictionary.index(z)] = c
            X[dictionary.index(z)][dictionary.index(k)] = c
        except ValueError:
            pass
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
S = cosine_similarity(X[dictionary.index('where')].reshape(1, -1), X)
```

```python
similar = [(i, s) for i, s in sorted(enumerate(S[0]), key=lambda x: -x[1])]
for i, s in similar:
    print(dictionary[i], s)
```
